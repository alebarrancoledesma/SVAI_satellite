### Visualización de resultados de la experimentación

```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
DATA_DIR = Path("./data")

dfs_clean = {}

for p in sorted(DATA_DIR.glob("*.csv")):
    df = pd.read_csv(p, skiprows=SKIPROWS)

    df = df.loc[:, ["x", "y"]].rename(columns={"x": "time", "y": "collisions"})

    df["time"] = pd.to_numeric(df["time"], errors="coerce")
    df["collisions"] = pd.to_numeric(df["collisions"], errors="coerce")

    df = df.dropna(subset=["time", "collisions"]).sort_values("time").reset_index(drop=True)

    dfs_clean[p.stem] = df
```

```python
df_r = dfs_clean[f"001 - RED"]
df_b = dfs_clean[f"001 - BLUE"]

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# RED
axes[0].plot(df_r["time"], df_r["collisions"])
axes[0].set_title(f"RED (no avoid)")
axes[0].set_xlabel("time")
axes[0].set_ylabel("collisions")
axes[0].set_xlim(0, 100000)
axes[0].set_ylim(0, 2500)
axes[0].grid(True)

# BLUE
axes[1].plot(df_b["time"], df_b["collisions"])
axes[1].set_title(f"BLUE (avoid)")
axes[1].set_xlabel("time")
axes[1].set_ylabel("collisions")
axes[1].set_xlim(0, 100000)
axes[1].set_ylim(0, 250)
axes[1].grid(True)

fig.savefig('./data/001 - plot.png', dpi=200, bbox_inches="tight")
fig.tight_layout()
plt.show()
```
